```python
import numpy as np

def load_binary_file(file_path):
    with open(file_path, 'rb') as f:
        # Read all the binary data
        binary_data = f.read()
        
        # Assuming each value in the binary file is 4 bytes (32 bits) representing a single precision floating-point number
        num_floats = len(binary_data) // 4
        
        # Convert the binary data to a NumPy array of floats
        float_array = np.frombuffer(binary_data, dtype=np.float32, count=num_floats)
        
        return float_array

# Example usage:
file_path = 'output.bin'  # Replace 'data.bin' with the path to your binary file
float_array = load_binary_file(file_path)
float_array = float_array.reshape(600,400*640)
print(float_array.shape)
float_array = float_array.reshape(600,400,640)

import scipy.io
scipy.io.savemat('output.mat', {'varr': float_array})
```
